```python
#codigo aula 03 - IMPORTAndo base de dados e gerando grafico
#importando as bibliotecas
import pandas as pd #permite a leitura do arquivo csv
import matplotlib.pyplot as plt #plota o grafico

```

```python
#cria uma variavel df para armazenar a base de dados
df = pd.read_csv("base_de_dados/monitoramento_tempo.csv",sep=",")
```

```python
df.head(15) #df head exibe as 4 primeiras linhas do dataset ou por quantidade escolhida, ex:15
```

```python
df.tail() #exibe as 4 ultimas linhas do dataset
```

```python
#verificando as informaçoes do dataset
df.info()
```

```python
#importa a biblioteac datetime

import datetime #biblioteca q permite transformar m objeto em datetime
df["data"] = pd.to_datetime(df["data"])#converte o objeto para um datetime
```

```python
df.info()
```

```python
df.describe() #exibe informações estatisticas da base de dados
```

```python
#criando o gráfico com a temperatura
plt.figure(figsize=(15,8))
plt.plot(df["data"],df["temperatura"])
plt.xlabel("data e hora",fontsize=18)
plt.ylabel("temperatura",fontsize=18)
```

```python
#cria variavel fig
fig = plt.figure(figsize=(10,8)) #cria varialvel fig para armazenar o tamanho da figura
eixo = fig.add_axes([0,0,1,1]) #define os eixos do grafico
eixo.plot(df["data"],df["temperatura"]/10,color="red",lw=0.2,linestyle="--")
eixo.legend(["temperatura"],loc="lower right",fontsize=15) #loc localização da legenda
eixo.set_xlabel("Ano",fontsize=20)
eixo.set_ylabel("temperatura",fontsize=20)
```

```python
#selecionando um periodo na base de dados
fig = plt.figure(figsize=(15,8))
plt.rcParams["font.size"]="20"
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df["data"],df["temperatura"],color="b",lw=0.5,linestyle="--")
#escolhendo um periodo da base de dados
eixo.set_xlim(datetime.datetime(2015,1,1),datetime.datetime(2015,12,31))
#valor inicial #valor final do periodo escolhido
eixo.set_title("Temperatura - Ano 2015",fontsize=25)
eixo.legend(["temperatura"],fontsize=20)
eixo.set_xlabel("Ano",fontsize=20)
plt.xticks(rotation=45,fontsize=15,ha="right")
plt.show()
```
