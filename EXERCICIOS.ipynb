```python
import pandas as pd #permite a leitura do arquivo csv
import matplotlib.pyplot as plt #plota o grafico
```

```python
#cria uma variavel df para armazenar a base de dados
df = pd.read_csv("base_de_dados/monitoramento_tempo.csv",sep=",")
```

```python
df.head(20)
```

```python
#importa a biblioteac datetime

import datetime #biblioteca q permite transformar m objeto em datetime
df["data"] = pd.to_datetime(df["data"])#converte o objeto para um datetime
```

```python
#criando o gráfico com a umidade
plt.figure(figsize=(12,6))
plt.title("GRÁFICO DE UMIDADE X ANO")
plt.plot(df["data"],df["humidade"])
plt.xlabel("Data e Hora",fontsize=15)
plt.ylabel("Umidade",fontsize=15)
```

```python
#criando o gráfico com a velocidade do vento
plt.figure(figsize=(12,6))
plt.plot(df["data"],df["velocidade do vento"])
plt.title("GRAFICO DE VELOCIDADE DO VENTO X ANO")
plt.xlabel("Data e Hora",fontsize=15)
plt.ylabel("Umidade",fontsize=15)
```

```python
#criando o gráfico com a umidade
plt.figure(figsize=(12,6))
plt.plot(df["data"],df["pressão"])
plt.title("GRAFICO DE VELOCIDADE DO VENTO X ANO")
plt.xlabel("Data e Hora",fontsize=15)
plt.ylabel("Pressão",fontsize=15)
```

```python
df = pd.read_csv('base_de_dados/monitoramento_tempo.csv')  # chamando arquivo csv para criação dos boxplots

# Plotar o boxplot da coluna 'umidade'
plt.figure(figsize=(6, 8))
plt.boxplot(df['umidade'].dropna())  # dropna para ignorar valores nulos
plt.title('Boxplot da Umidade')
plt.ylabel('Umidade')
plt.show()
```

```python
# Plotar o boxplot da coluna 'velocidade do vento'
plt.figure(figsize=(6, 8))
plt.boxplot(df['velocidade do vento'].dropna())  # dropna para ignorar valores nulos
plt.title('Boxplot do Vento')
plt.ylabel('Velocidade do vento')
plt.show()
```

```python
# Plotar o boxplot da coluna 'pressao'
plt.figure(figsize=(6, 8))
plt.boxplot(df['pressão'].dropna())  # dropna para ignorar valores nulos
plt.title('Boxplot da Pressão')
plt.ylabel('Pressão')
plt.show()
```
