```python
#importa  as biblioteca
import matplotlib.pyplot as plt  
import numpy as np
```

```python
#gerando dados para plotar o grafico
#linspace gera valores espaçados de 10 em 10
Dados_x = np.linspace(0,10,10)
Dados_y = Dados_x * Dados_x
```

```python
plt.figure(figsize=(10,8),facecolor="gray") #define o tamanho e a cor do fundo da figura
#Cria a figura
Ax = plt.axes() #cria os eixos do grafico
Ax.set_facecolor("white") #cor do fuundo do grafico
plt.plot(Dados_x,Dados_y,color="red",marker="o",linestyle="dashed",markeredgecolor="black")
plt.xlabel("Eixo_x",fontsize=16) #titulo do eixo x
plt.xlabel("Eixo_y", fontsize=16) #titulo do eixo y
plt.title("Gráfico simples com matplotlib",fontsize=16) #titulo do grafico
```

```python
#subplot

plt.rcParams.update({"font.size":15})
plt.figure(figsize=(12,12))
linhas = 2
colunas =2
#subplot para plotar figuras
plt.subplot(linhas,colunas,1)
plt.plot(Dados_x,"r--",label ="Dados_x")
plt.legend(loc="lower right")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Gráfico 1")


#grafico 2

plt.subplot(linhas,colunas,2)
plt.plot(Dados_x,Dados_y,"b--",label="Dados x \n e y")
plt.title("Gráfico 2")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.legend()

#grafico 3

plt.subplot(linhas,colunas,3)
plt.plot(Dados_x*Dados_x,"g--",label="Dados_x^2")
plt.title("Gráfico 3")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.legend()

#grafico 4

plt.subplot(linhas,colunas,4)
plt.plot(Dados_y,"c--",label="Dados_y")
plt.title("Gráfico 4")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.legend()

plt.savefig("Gráfcos aula 3.png")
plt.show() #remove warning


```

```python
# criando grafico de barras

Figura =plt.figure()
Grafico = Figura.add_axes([0,0,0.7,0.7])
#cria rotulos
Rotulos =["A","B","C","D","E","F","G","H","I","J"]
Grafico.bar(Rotulos,Dados_x,color="red")
Grafico.set_xlabel("Rotulos")
Grafico.set_ylabel("Quantidade")
```

```python
#Criando Grafico de barras azul e vermelho
Grafico.bar(Rotulos,Dados_x,color="red") #dados 1
Grafico.bar(Rotulos,Dados_y*0,1,color="blue") #dados 2
Figura
```

```python
Grafico.bar(Dados_x,Dados_y/2,color=["red","blue"])
Figura
```

```python
#plotando boxplot

plt.boxplot(Dados_x) #plota o boxplot
plt.grid()
plt.show()
```
